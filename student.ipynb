```python
import pandas as pd

try:
    df = pd.read_csv('/content/AviationData.csv',low_memory = False,encoding='latin-1')
    display(df.head())
    display(df.info())
except FileNotFoundError:
    print("Error: 'Aviation_Data.csv' not found. Please ensure the file exists in the current directory.")
    df = None
```

```python
# Examine the shape of the DataFrame
print(f"DataFrame Shape: {df.shape}")

# Re-examine data types and non-null values
df.info()

# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().sum() * 100 / len(df)
print("\nPercentage of Missing Values:")
print(missing_percentage)

# Summarize initial findings (This will be printed to the console for now)
print("\nInitial Data Exploration Summary:")
print(f"- The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
print("- Data types and non-null value counts have been re-examined using .info().")
print("- Missing value percentages have been calculated.")
print("- Further analysis is needed to determine numerical and categorical features and handle missing values.")
```

```python
#Data Analysis
import matplotlib.pyplot as plt

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Descriptive statistics for numerical features
print("Descriptive Statistics for Numerical Features:")
display(df[numerical_cols].describe())

# Analyze distribution of categorical features
print("\nFrequency Counts and Unique Values for Categorical Features:")
for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(f"Unique values: {df[col].unique()}")
    print(f"Value counts:\n{df[col].value_counts()}")

    # Plot value counts for selected categorical columns (top 5)
    if col in ['Investigation.Type', 'Aircraft.Category', 'Weather.Condition', 'Broad.phase.of.flight', 'Purpose.of.flight']:
        plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
        df[col].value_counts().plot(kind='bar', color='skyblue')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
        plt.tight_layout()
        plt.show()
```

```python
#Data Visualization
import matplotlib.pyplot as plt

# Histograms for Numerical Features
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    if col not in ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']:  # Exclude injury counts
        plt.figure(figsize=(8, 6))
        df[col].hist(bins=20, color='skyblue', edgecolor='black')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.show()

# Bar Charts for Categorical Features
categorical_cols = ['Investigation.Type', 'Aircraft.Category', 'Weather.Condition', 'Broad.phase.of.flight']
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Scatter Plot and Box Plots for Relationships
plt.figure(figsize=(8, 6))
plt.scatter(df['Number.of.Engines'], df['Total.Fatal.Injuries'], alpha=0.5, color='green')
plt.title('Number of Engines vs. Total Fatal Injuries')
plt.xlabel('Number of Engines')
plt.ylabel('Total Fatal Injuries')
plt.show()

plt.figure(figsize=(8,6))
df.boxplot(column='Total.Fatal.Injuries', by='Aircraft.Category',figsize=(10,6))
plt.title('Total Fatal Injuries by Aircraft Category')
plt.suptitle('') # Suppress default title
plt.xlabel('Aircraft Category')
plt.ylabel('Total Fatal Injuries')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
